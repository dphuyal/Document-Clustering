```python
import numpy as np

total_words = []
for i in range(10):
    
    doc = []

    word1 = [1, 2, 3]
    word2 = [4, 5, 6]
    word3 = [7, 8, 9]

    doc.append(word1)
    doc.append(word2)
    doc.append(word3)
    
    doc = np.mean(doc, axis = 0)

    
    total_words.append(doc)
```

```python
print('Shape: ', np.shape(total_words))
for doc in total_words:
    print(doc)
```
